```python
import sys
import pkg_resources
from importlib.metadata import version, distributions

# Display Python version
print(f"Python Version: {sys.version}")
print(f"Python Executable: {sys.executable}")
print("\n" + "="*70)
print("INSTALLED PACKAGES AND VERSIONS")
print("="*70 + "\n")

# Get all installed packages sorted by name
packages = sorted([(d.metadata['Name'], d.metadata['Version']) for d in distributions()], key=lambda x: x[0].lower())

print(f"Total packages installed: {len(packages)}\n")
for pkg_name, pkg_version in packages:
    print(f"{pkg_name:<40} {pkg_version}")
```
